```python
def add_signals(env):
    start = env.frame_bound[0] - env.window_size
    end = env.frame_bound[1]
    prices = env.df.loc[:, 'Low'].to_numpy()[start:end]
    signal_features = env.df.loc[:, ['Low', 'Volume','SMA', 'RSI', 'OBV']].to_numpy()[start:end]
    
    return prices, signal_features
```

```python

# Custom environment
class CustomForexEnv(ForexEnv):
    def __init__(self, df, window_size, frame_bound, unit_side='left', render_mode=None):
        super().__init__(df, window_size, frame_bound, unit_side, render_mode)
        self.trade_fee = 0.03  # Updated trade fee

    def _calculate_reward(self, action):
        step_reward = 0  # pip

        trade = False
        if (
            (action == Actions.Buy.value and self._position == Positions.Short) or
            (action == Actions.Sell.value and self._position == Positions.Long)
        ):
            trade = True

        if trade:
            current_price = self.prices[self._current_tick]
            last_trade_price = self.prices[self._last_trade_tick]
            price_diff = current_price - last_trade_price

            if self._position == Positions.Short:
                step_reward += -price_diff * 100  # Updated reward calculation
            elif self._position == Positions.Long:
                step_reward += price_diff * 100  # Updated reward calculation

        return step_reward

    _process_data = add_signals
```
